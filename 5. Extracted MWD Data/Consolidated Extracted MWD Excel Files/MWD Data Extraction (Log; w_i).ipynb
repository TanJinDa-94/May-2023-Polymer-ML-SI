```python
df = pd.read_excel(file_list[0]) # Read 1st file

df # DataFrame needs cleaning to extract MW and Mole Fraction data
```

```python
x = list(df.iloc[:,11].dropna()) # Drop all empty values
M_i = x[1:] # Drop 1st element 'Molecular Weight' of list

w_i = list(df.iloc[:,13].dropna()) # Drop all empty values
w_i = w_i[1:] # Drop 1st element 'Normalized Weight Fraction' of list. w_i list represents the Normalized Mole Fraction.

'''
Create the scaffold of a DataFrame so that each file can be iteratively read and then appended in rows.
The DataFrame will contain the absolute molecular weights as its header, and the values of w_i in each file as its values.
The length of each row is fixed to 1499 values, owing to the boundary set during data extraction using OmniSEC.
Each row will represent an MWD when plotted against the header.
'''

df = dict(zip(M_i,w_i)) # zip the M_i and w_i lists together and create a dictionary
df = pd.DataFrame(df, index=[0]) # Create a Dataframe from the dictionary and set index to log MW.
df # Inspect df. Header must be the log molecular weight, each row represents the w_i value of each file at that MW.
```

```python
for i in range(len(file_list)): # i ranges from 0 to 107 (108 MWD files)
    df1 = pd.read_excel(file_list[i]) # Dummy df1 object stores the unedited excel file DataFrame within the loop
    w_i = list(df1.iloc[:,13].dropna()) # Drop missing values and create a list
    w_i = w_i[1:] # Drop 'Normalized Wt Fraction'
    df.loc[i] = w_i # Overwrite the ith row of df with w_i

df = df.div(df.sum(axis=1), axis=0)

# Inspect df
df
```

```python
'''
Re-index the rows of df into the corresponding condition and residence time for book-keeping purposes.
'''

index =[]
for key, value in file_list.items():
    value = value.replace('.xlsx','') # Replace substring '.xlsx' with ''
    index.append(value) 

df.index = index
df
```

```python
l = np.arange(0,105,4) # Create a list which encodes the index location for every 4 rows.


for k,i in enumerate(l):
    
    df_dummy = df.iloc[i:i+4,:]
    x = df.T.index
    y1 = df.iloc[i,:]
    y2 = df.iloc[i+1,:]
    y3 = df.iloc[i+2,:]
    y4 = df.iloc[i+3,:]
    
    plt.figure(figsize=(9,5))
    plt.title('Condition ' + str(k+1)+ ' Overlay', fontsize=16)
    plt.plot(x, y1, label='30 min')
    plt.plot(x, y2, label='60 min')
    plt.plot(x, y3, label='90 min')
    plt.plot(x, y4, label='120 min')
    plt.xlabel('Log Molecular Weight',size=16)
    plt.ylabel('Normalized Weight Fraction ($w_{i}$)',size=16)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.legend(fontsize=16)
    plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
    plt.show()
```
